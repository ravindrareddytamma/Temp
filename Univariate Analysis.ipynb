### Visualizing a Categorical Column

```python
### bar plot for department wise count
data.sales.value_counts().plot.bar()

### Most of the people in the dataset are from sales department
### Management has less number of people
```

```python
### bar plot with percentages
round(data.sales.value_counts()/len(data.sales)*100,2).plot.bar()

# Sales contributes about 28% of total population
# sales,Technical,Support,IT combinedly contributes around 50% population
```

# Visualizing a Ordinal Column

```python
### bar plot for a Ordnial Column
data.time_spend_company.value_counts().plot.bar()

## Almost 6300 members handeld 3 Projects
## Almost 91% of people handle less than 5 Projects
```

```python
temp = data.time_spend_company.value_counts().rename_axis('Time').reset_index(name = 'Count')
temp['CumCount'] = temp.Count.cumsum()
temp['CumCount'] = round(temp.CumCount*100/14999,2)
temp.set_index('Time',drop=False,inplace=True)
```

```python
fig, ax = plt.subplots(2, 1,sharex=True)

temp.plot(x='Time', y='Count', kind='bar', figsize=(15, 10), legend=False, style='yo-',ax=ax[1])
temp.plot(x='Time', y='CumCount', kind='line', figsize=(15, 10), legend=False,
          style='mo-',ax=ax[0],use_index = False)
```
