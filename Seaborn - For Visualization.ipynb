```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Cars.csv')
data.head(5)
# data.columns
```

# Scatter Plot

```python
sns.scatterplot(x = 'Price',y = 'MPG.city',data = data)
plt.show()

### Low priced cars have high mileage
### High Priced cars have less mileage
```

# Count Plot

```python
sns.countplot(x = 'AirBags',data = data)
```

# Adding third variable as Color

```python
sns.scatterplot(x = "Length", y = 'Wheelbase',data = data)
```

```python
## Adding hue/color for third variable
sns.scatterplot(x = "Length", y = 'Wheelbase',data = data,hue = 'Type')
```

```python
hue_colors = {'Small' : 'Red','Midsize':'Green','Compact':'Blue','Large':'Yellow','Sporty':'Cyan','Van':'black'}
sns.scatterplot(x = "Length", y = 'Wheelbase',data = data,hue = 'Type',palette=hue_colors)
```

```python
## Countplot with hue
sns.countplot(x = "Type",hue = 'Origin',data = data,palette={'USA':'#FF8C00','non-USA':'#01173D'})

## All large vehicles are from USA
## Most of the small vehicles are from Non USA
## USA is famous for Large,Sporty & Vans
```

# Relational Plots

### Relational Plots are used to find the relations between 2 Quantitative variables

#### 1. Relplots helps to find the relations seperately in 2 plot figures
#### 2. They use both scatter & line graphs for quantitative variables

```python
### Simple Scatter plot
sns.relplot(x = "Length", y = 'Wheelbase',data = data,kind = 'scatter')
```

```python
### Two Scatter plots arrange in columns
sns.relplot(x = "Length", y = 'Wheelbase',data = data,kind = 'scatter',col = 'Origin')
```

```python
### Two Scatter plots arrange in Rows
sns.relplot(x = "Length", y = 'Wheelbase',data = data,kind = 'scatter',row = 'Origin')
```

```python
### Two Scatter plots arranged in Rows & Colums
sns.relplot(x = "Length", y = 'Wheelbase',data = data,kind = 'scatter',row = 'Origin',col = 'AirBags')
```

# Adding more dimensions to Scatter Plot

### More dimensions can be added to the plots using below parameters
#### 1. Size
#### 2. Shape
#### 3. Transperency

```python
# Size
sns.relplot(x = 'Passengers', y = 'Weight',size='AirBags',data = data,hue='AirBags',kind='scatter')

### High Weight cars with 6 & 7 Passengers has higher saftey
```

```python
# Shape / style
sns.relplot(x = 'Passengers', y = 'Weight',style='Origin',data = data,hue='Origin',kind='scatter')
```
